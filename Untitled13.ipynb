```python
# Data manipulation and analysis
import pandas as pd
import numpy as np
```

```python
# Define the file path
file_path = "C:\\Users\\HP\\Downloads\\H.R_Analysis_project.csv"

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
df.head()
```

```python
# Identify duplicate rows
duplicate_rows = df[df.duplicated()]

# Display duplicate rows
print("Duplicate Rows:")
print(duplicate_rows)

# Count duplicate rows
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {num_duplicates}")

#  Remove rows with any NaN values
df_cleaned_rows = df.dropna()

# Display the first few rows of the cleaned DataFrame
df_cleaned_rows.head(10)
```

```python
# Using shape attribute determine the number of columns
num_columns_shape = df.shape[1]
print(f"Number of columns (using shape): {num_columns_shape}")
```

```python
# Get the number of non-null rows for each column
row_counts = df.count()

# Print the number of rows in each column
for column, count in row_counts.items():
    print(f"Column '{column}': {count} non-null rows")
```
